```python
#Abrimos el archive y lo llamamos file1
file1=open('Starwars.txt')
```

```python
#Leemos el archivo y lo pasamos a la variable file
file=file1.read()
file=file.split()
file2=[]
```

```python
#Una vez definida file2 procedemos a llenarla con el contenido de file
for i in file:
    if i not in file2:
        file2.append(i)
```

```python
#Comparamos el número de letras de cada una de las palabras de file2 para imprimir solo las que tienen más de cuatro letras.
for i in range(0, len(file2)):
    if(len(file2[i]))>4:
        print('(',file2[i],',', file.count(file2[i]),')')
```
